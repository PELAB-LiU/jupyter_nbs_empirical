# Parse ast for error notebooks

## 1. Parse Kaggle error notebooks into ast

```python
import config
import ast_parser

path_errnbs_kaggle = config.path_kaggle_error_process + "/k_error_nbs"
res_fasts = ast_parser.parse_nbs_to_asts(path_errnbs_kaggle, config.path_kaggle_error_ast, ast_parser.parse_code_ast)
```

```python
res_fasts.to_excel(config.path_kaggle_error_process + "/nberror_k_ast.xlsx", index=False, engine='xlsxwriter')
```

```python
# how many percent can be parsed
print("Number of error notebooks that can be parsed to AST: {}({:.2%})".format(len(res_fasts),len(res_fasts)/4331))
```

```python
# test saved asts
import pickle
import ast

with open(config.path_kaggle_error_ast+"/"+res_fasts.iloc[0]['fast'], 'rb') as f:
    root = pickle.load(f)
    for node in ast.walk(root):
            res_import = {"from":[], "import":[], "as":[]}
            if isinstance(node, ast.Import):
                pass
            elif isinstance(node, ast.ImportFrom):  
                res_import["from"] = node.module.split('.')
            else:
                continue

            for n in node.names:
                res_import["import"] = n.name.split('.')
                res_import["as"] = n.asname.split('.') if n.asname!=None else []
                print(res_import)
```

## 2. Check python version distribution of Kaggle and GitHub error notebooks

```python
# kaggle
```

```python
import config
import ast_parser

path_errnbs_kaggle = config.path_kaggle_error_process + "/k_error_nbs"
res_k = ast_parser.nb_python_version_exact(path_errnbs_kaggle)
```

```python
res_k.to_excel(config.path_kaggle_error_process + "/nberror_k_pyversion.xlsx", index=False, engine='xlsxwriter')
res_k.python_version.value_counts() # all are python 3
```

```python
# github
```

```python
path_errnbs_github = config.path_github_error_process + "/nbs"
res_g = ast_parser.nb_python_version_exact(path_errnbs_github)
```

```python
import pandas as pd

df_err_g = pd.read_excel(config.path_github_error_process + "/nberror_g_all_p.xlsx")
```

```python
res_g_mapped = pd.merge(df_err_g, res_g, on="fname", how="left")[["fname","python_version"]].drop_duplicates()
res_g_mapped["python_version"] = res_g_mapped.python_version.apply(lambda pv: int(str(pv)[0]) if pv/10>1 else pv)
res_g_mapped.to_excel(config.path_github_error_process + "/nberror_g_pyversion.xlsx", index=False, engine='xlsxwriter')
res_g_mapped.python_version.value_counts()
```

## 3. Parse GitHub error notebooks into ast

Because github error notebooks contain both python 2 and python 3 versions, we use [tree-sitter](https://github.com/tree-sitter/py-tree-sitter). But tree-sitter cannot be serialized directly, so the ast results cannot be saved as the python built-in ast module.

```python
import config
import ast_parser

path_errnbs_github = config.path_github_error_process + "/nbs"
parser = ast_parser.py23_setup_parser()
res_fasts = ast_parser.parse_nbs_to_asts(path_errnbs_github, None, ast_parser.parse_code_ast_py23, parser=parser) # cannot save ast (not support serialization)
```

```python
res_fasts.to_excel(config.path_github_error_process + "/nberror_g_all_ast.xlsx", index=False, engine='xlsxwriter')
```
