## 3. Combine the results from metalanguage exaction and language detection


```python
import pandas as pd
import util

path = r'C:\Users\yirwa29\Downloads\Dataset-Nb\nbdata_g'

res_lan1 = pd.read_excel(path+'/nbdata_g_metalanguage_0-99.xlsx')
res_lan2 = pd.read_excel(path+'/nbdata_g_metalanguage_100-199.xlsx')
res_lan3 = pd.read_excel(path+'/nbdata_g_metalanguage_200-311.xlsx')

res_lan_detect = pd.read_excel(path+'/nbdata_g_detectlanguage.xlsx')
res_lan_detect['language'] = res_lan_detect.language.map(str.lower)
```

```python
res_lan = pd.concat([res_lan1, res_lan2, res_lan3], ignore_index=True)
```

```python
df_tmp = res_lan[res_lan['language'].eq('unknown')].merge(res_lan2, on='fname', how='left')
for i, row in df_tmp.iterrows():
    res_lan.loc[res_lan['fname']==row['fname'], 'language'] = row['language_y']
```

## 4. Language distribution in GitHub notebooks

```python
import pandas as pd

path = r'C:\Users\yirwa29\Downloads\Dataset-Nb\nbdata_g'

res_lan1 = pd.read_excel(path+'/nbdata_g_language_1.xlsx')
res_lan2 = pd.read_excel(path+'/nbdata_g_language_2.xlsx')
res_lan = pd.concat([res_lan1, res_lan2], ignore_index=True)
```

```python
res_lan.language.value_counts()[:3]/len(res_lan)
```

```python
res_lan[res_lan.language=="python"].to_excel(path+'/nbdata_g_language_python.xlsx', index=False, engine='xlsxwriter')
```

```python
# tried to plot, looks ugly. May just use table

import matplotlib.pyplot as plt
import seaborn as sns 

#the full dataframe
lan_df = pd.DataFrame(
    data = {'language': res_lan.language.value_counts().index, 
            'value' :res_lan.language.value_counts().values},
    ).sort_values('value', ascending = False)

#the top n
n_top = 3
lan_df_1 = lan_df[:n_top].copy()

#others
lan_df_2 = pd.DataFrame(data = {
    'language' : ['others'],
    'value' : [lan_df['value'][n_top:].sum()]
})

# combine
lan_df_new = pd.concat([lan_df_1, lan_df_2])

explode = (0.1, 0, 0, 0)

plt.figure(figsize=(4,4))
plt.tight_layout()
plt.pie(
    x = lan_df_new['value'], 
    explode = explode, 
    labels = lan_df_new['language'],
    autopct = '%1.2f%%',
    colors=sns.color_palette('Set2')
)
plt.show()
```
