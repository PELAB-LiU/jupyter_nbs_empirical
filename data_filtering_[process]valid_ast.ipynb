# Use the py3 AST module to filter invalid notebooks

Invalid notebooks are notebooks with syntax errors or using older Python versions, making them invalid to Python3 grammar.

## 1. Parse Kaggle error notebooks into ast

```python
import utils.config as config
import utils.ast_parser as ast_parser

path_errnbs_kaggle = config.path_kaggle_error_process + "/k_error_nbs"
res_fasts = ast_parser.parse_nbs_to_asts(path_errnbs_kaggle, config.path_kaggle_error_ast, ast_parser.parse_code_ast)
```

```python
res_fasts.to_excel(config.path_kaggle_error_process + "/nberror_k_ast.xlsx", index=False, engine='xlsxwriter')
```

```python
# how many percent can be parsed
print("Number of error notebooks that can be parsed to AST: {}({:.2%})".format(len(res_fasts),len(res_fasts)/4331))
```

```python
# test saved asts
import pickle
import ast

with open(config.path_kaggle_error_ast+"/"+res_fasts.iloc[0]['fast'], 'rb') as f:
    root = pickle.load(f)
    for node in ast.walk(root):
            res_import = {"from":[], "import":[], "as":[]}
            if isinstance(node, ast.Import):
                pass
            elif isinstance(node, ast.ImportFrom):  
                res_import["from"] = node.module.split('.')
            else:
                continue

            for n in node.names:
                res_import["import"] = n.name.split('.')
                res_import["as"] = n.asname.split('.') if n.asname!=None else []
                print(res_import)
```
