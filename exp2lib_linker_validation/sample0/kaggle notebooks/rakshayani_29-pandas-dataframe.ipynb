```python
d = pd.DataFrame(d)
d
```

```python
d.dropna(how = 'all',inplace = True)
d
```

```python
d.dropna(axis = 0, subset = ['Age'])
```

```python
d.dropna(axis = 1,subset = [1])
```

# **Ranking** the data using rank() functions

```python
import pandas as pd
```

```python
c={
   'Country': ['India','China','Japan','Bangladesh','Singapore'],
    'Gold': [12,20,11,3,None]
}
c = pd.DataFrame(c)
c
```

```python
c['Rank']=c['Gold'].rank(ascending=True)
c.sort_values('Rank',inplace=True)
c
```

**With na**

```python
c['Rank']=c['Gold'].rank(ascending=False, na_option='bottom')
c
```

**What happens when data is same?**

```python
c={
   'Country': ['India','China','Japan','Bangladesh','Singapore'],
    'Gold': [12,20,12,3,None]
}
c = pd.DataFrame(c)
c
```

```python
c['Rank']=c['Gold'].rank(ascending=False, na_option='bottom')
c.sort_values('Rank',inplace=True)
c
```

**Change aggregations**

```python
a = pd.read_csv('../input/videogamesales/vgsales.csv')
a['Rank']=a['Global_Sales'].rank(ascending=False)
a
```

```python
a = pd.read_csv('../input/videogamesales/vgsales.csv')
a['Rank']=a['Other_Sales'].rank(ascending=False)
a.sort_values('Rank').head()
```
