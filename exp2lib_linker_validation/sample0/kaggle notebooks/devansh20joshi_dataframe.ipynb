**Applying a transformation across all the values of certain columns**

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/kaggle/input/data-science-jobs-in-india/naukri_data_science_jobs_india.csv")
data.head()
```

```python
data["Salary"] = np.random.randint(400000,4000000,12000)
```

```python
data.head()
```

```python
data["Salary"] = data["Salary"].apply(lambda x:"Rs" + str(x))
```

```python
data
```

```python
def band(x):
    if x > 2000000:
        return "A"
    elif x > 1000000:
        return "B"
    else:
        return "C"
```

```python
data["category"] = data["Salary"].apply(band)
```

```python
data
```

**Applying the transformation to a particulare row**

```python
def flage(x):
    loc = x["Location"]
     
    exp = x["Job Experience"]
    if loc in ["Bengaluru","Bangalore","Bangalore/Bengaluru"] and exp == "3-5":
        return "Yes"
    else:
        return "no"
```

```python
data["Apply Or Not"] = data.apply(flage,axis = 1)
```

```python
data["Apply Or Not"].value_counts()
```
