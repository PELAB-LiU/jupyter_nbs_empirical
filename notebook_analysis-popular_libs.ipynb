## 2. Analysis and sort

Analysis and sort imported library information extracted from the ENTIRE Kaggle notebooks

```python
import pandas as pd
import config

path_imports = config.path_default + "/nbdata_k/nbdata_k_imports.xlsx"
imports_pd = pd.read_excel(path_imports)
imports_pd.head()
```

```python
imports_pd['imports'] = imports_pd['imports'].apply(eval)
nf_lib = imports_pd[imports_pd["imports"].str.len()!=0]
print("Successfully processed {0} Kaggle notebooks, among which {1:.2%}({2}) used imported libraries.".format(len(imports_pd), 
                                                                                                           len(nf_lib)/len(imports_pd),
                                                                                                           len(nf_lib)))

imports_pd_reversed = imports_pd.explode('imports', ignore_index=True)
# doing this because should extend the list but did append list when adding multiple imports from the same statement
imports_pd_reversed["imports2"] = imports_pd_reversed.imports.apply(lambda imps: [imp.strip() for imp in str(imps).split(',')] if isinstance(imps,str) else imps)
imports_pd_reversed2 = imports_pd_reversed.explode('imports2', ignore_index=True)
imports_pd_reversed2 = imports_pd_reversed2.drop(columns=["imports"]).rename(columns={"imports2":"imports"}).drop_duplicates()

n_libs = 50
imports_counts = imports_pd_reversed2.imports.value_counts()
print("The top {0} libraries ({1:.2%}) are: \n {2}".format(n_libs, sum(imports_counts[:n_libs])/sum(imports_counts), imports_counts[:n_libs]))
```

```python
list_libs = imports_counts[:n_libs].index.to_list()
n_nb_libs = len(set(imports_pd_reversed2[imports_pd_reversed2.imports.isin(list_libs)].fname.to_list()))
nb_coverage = n_nb_libs/len(set(imports_pd.fname.to_list())) # 166654, not only the ones used libs
print("The top {0} libraries are used in {1}({2:.2%}) of all the processed notebooks".format(n_libs, n_nb_libs, nb_coverage))
```

```python
pd.DataFrame.from_dict(imports_counts).to_excel(config.path_default + "/nbdata_k/nbdata_k_popular_libs.xlsx", engine='xlsxwriter')
```

## 4. Check if the static import statements are actually used in the code

```python
import imports_parser
import pandas as pd
import config

# take a sample
# path_imports = config.path_default + "/nbdata_k/nbdata_k_imports.xlsx"
# imports_pd = pd.read_excel(path_imports)
# imports_pd_sample = imports_pd.sample(n=int(0.1*len(imports_pd))) # sample 10%
# imports_pd_sample.to_excel(config.path_default + "/nbdata_k/nbdata_k_imports_sample.xlsx", index=False, engine='xlsxwriter')

imports_pd_sample = pd.read_excel(config.path_default + "/nbdata_k/nbdata_k_imports_sample.xlsx")
```

```python
# get the sampled nbs
import os, shutil

path_samples = config.path_drive_kaggle + "/nbdata_k_sample"
fnames = list(imports_pd_sample.fname)
for path, _, files in os.walk(config.path_drive_kaggle+"/nbdata_k"):
    for file in files:
        if file.endswith(".ipynb") and (file in fnames):
            src_file = f"{path}/{file}"
            dst_file = f"{path_samples}/{file}"
            if os.path.exists(dst_file):
                if os.path.samefile(src_file, dst_file):
                    continue
                os.remove(dst_file)
            shutil.copy(src_file, dst_file)
```

```python
# extract imports of the sampled nbs by ast
path_samples = config.path_drive_kaggle + "/nbdata_k_sample"
# get_imports_pys_outermost_ast   # if all are pys
res_sample_ast = imports_parser.get_imports_nbs_outermost_ast(path_samples, get_imports_func = imports_parser.get_imports_py_outermost)
imports_pd_sample_ast = pd.DataFrame.from_dict(res_sample_ast)
imports_pd_sample_ast.to_excel(config.path_default + "/nbdata_k/nbdata_k_imports_ast_samples.xlsx", index=False, engine='xlsxwriter')
```

```python
# compare the static extraction and ast extraction of imports on the sampled nbs

imports_pd_sample = pd.read_excel(config.path_default + "/nbdata_k/nbdata_k_imports_sample.xlsx") # static
imports_pd_sample_ast = pd.read_excel(config.path_default + "/nbdata_k/nbdata_k_imports_ast_samples.xlsx") # ast

compare_df = pd.merge(imports_pd_sample_ast, imports_pd_sample, on="fname",how="left")
compare_df["not_used"] = ""
compare_df["count_not_used"] = ""
compare_df["count_static"] = ""
compare_df["count_ast"] = ""

for index, row in compare_df.iterrows():
    set_static = eval(row.imports_y)
    set_ast = eval(row.imports_x)
    compare_df.loc[index, "not_used"]=str(set_static-set_ast)
    compare_df.loc[index, "count_not_used"]=len(set_static-set_ast)
    compare_df.loc[index, "count_static"]=len(set_static)
    compare_df.loc[index, "count_ast"]=len(set_ast)
    
compare_df.to_excel(config.path_default + "/nbdata_k/nbdata_k_imports_sample_compare.xlsx")
```

```python
print("There exists {0:.2%} notebooks have certain unused imports".format(len(compare_df[compare_df.count_not_used>0])/len(compare_df))) # notebook
print("There exists {0:.2%} imports are unused".format(sum(compare_df.count_not_used)/sum(compare_df.count_static))) # import
```
