## 2. Analysis and sort

Analysis and sort imported library information extracted from the ENTIRE Kaggle notebooks

```python
import pandas as pd

path_imports = "C:/Users/yirwa29/Downloads/Dataset-Nb/nbdata_k/nbdata_k_imports.xlsx"
imports_pd = pd.read_excel(path_imports)
imports_pd.head()
```

```python
imports_pd['imports'] = imports_pd['imports'].apply(eval)
nf_lib = imports_pd[imports_pd["imports"].str.len()!=0]
print("Successfully processed {0} Kaggle notebooks, among which {1:.2%}({2}) used imported libraries.".format(len(imports_pd), 
                                                                                                           len(nf_lib)/len(imports_pd),
                                                                                                           len(nf_lib)))

imports_pd_reversed = imports_pd.explode('imports', ignore_index=True)
# doing this because should extend the list but did append list when adding multiple imports from the same statement
imports_pd_reversed["imports2"] = imports_pd_reversed.imports.apply(lambda imps: [imp.strip() for imp in str(imps).split(',')] if isinstance(imps,str) else imps)
imports_pd_reversed2 = imports_pd_reversed.explode('imports2', ignore_index=True)
imports_pd_reversed2 = imports_pd_reversed2.drop(columns=["imports"]).rename(columns={"imports2":"imports"}).drop_duplicates()

n_libs = 50
imports_counts = imports_pd_reversed2.imports.value_counts()
print("The top {0} libraries ({1:.2%}) are: \n {2}".format(n_libs, sum(imports_counts[:n_libs])/sum(imports_counts), imports_counts[:n_libs]))
```

```python
list_libs = imports_counts[:n_libs].index.to_list()
n_nb_libs = len(set(imports_pd_reversed2[imports_pd_reversed2.imports.isin(list_libs)].fname.to_list()))
nb_coverage = n_nb_libs/len(set(imports_pd.fname.to_list())) # 166654, not only the ones used libs
print("The top {0} libraries are used in {1}({2:.2%}) of all the processed notebooks".format(n_libs, n_nb_libs, nb_coverage))
```
