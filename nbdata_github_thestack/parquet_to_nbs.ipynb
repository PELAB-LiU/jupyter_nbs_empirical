## Convert all the 'parquet' datafiles to notebooks

total number of notebooks in the original parquet files: 1073534

total number of successfully decoded notebooks from the original parquet files: 1070293 (3241 could not convert to json)

    parquet 0-99: 343098 nb files
    parquet 100-199: 343030 nb files
    parquet 200-311: 384165 nb files

```python
import os
import json
import shutil
import pandas as pd
from slugify import slugify
from os import listdir
from os.path import isfile

#datapath = 'C:/Users/yirwa29/Downloads/Dataset-Nb/nbdata_g/datafiles'
nbpath = 'C:/Users/yirwa29/Downloads/Dataset-Nb/nbdata_g/nbfiles'
#nbpath = 'P:/Downloads/tmp'

datapath = 'P:/Downloads/nbdata_github_source'

n_decoding_error = 0

for i in range(200, 312):
    id_name = str(i).zfill(5)
    filepath = os.path.join(datapath, 'data-{}-of-00312.parquet'.format(id_name))
    
    savepath = os.path.join(nbpath, id_name)
    if not os.path.exists(savepath):
        os.makedirs(savepath)
    
    if os.path.isfile(filepath):
        df = pd.read_parquet(filepath)
        
        for index, row in df.iterrows():
            file_name = id_name+'_'+str(index)+'_'+row['max_stars_repo_path'].split('/')[-1].replace(' ', '_').replace('.ipynb', '')
            file_name = slugify(file_name)
            file_name = file_name + '.' + row['ext']
            c = row['content']
            try:
                file_as_json = json.loads(c)
                with open(os.path.join(savepath, file_name), 'w', encoding='utf-8') as f:
                    json.dump(file_as_json, f, ensure_ascii=False, indent=4)
            except json.decoder.JSONDecodeError:
                n_decoding_error += 1
                print("decoding error: {}".format(file_name))
print("\n total number of notebooks that cannot be decoded: {}".format(n_decoding_error))
```

```python
# how many nbs in total are obtained
sum([len(files) for r, d, files in os.walk(nbpath)])
```

```python
n_nbs = 0

for i in range(200, 312):
    id_name = str(i).zfill(5)
    filepath = os.path.join(datapath, 'data-{}-of-00312.parquet'.format(id_name))
    
    if os.path.isfile(filepath):
        df = pd.read_parquet(filepath)
        n_nbs += df.shape[0]
print("total number of notebooks in the original parquet files: {}".format(n_nbs))
```

## Download all the parquet files

https://huggingface.co/datasets/bigcode/the-stack-dedup/tree/main/data/jupyter-notebook

```python
from urllib import request
import os
import requests

datapath = 'P:/Downloads/nbdata_github_source'

for i in range(181, 312, 1):
    id_name = str(i).zfill(5)
    remote_url = 'https://huggingface.co/datasets/bigcode/the-stack-dedup/resolve/main/data/jupyter-notebook/data-{}-of-00312.parquet?download=true'.format(id_name)
    local_file = 'data-{}-of-00312.parquet'.format(id_name)
    local_file_path = os.path.join(datapath, local_file)
    
    data = requests.get(remote_url, headers={'Authorization': ''})  
    if data.ok:
        with open(local_file_path, 'wb')as file:
            file.write(data.content)
    else:
        print('downloading failed: {}'.format(data.status_code))
```
